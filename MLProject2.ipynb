```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# 1️⃣ Define data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Hours
Y = np.array([30, 50, 65, 70, 85])            # Marks

# 2️⃣ Create polynomial features (degree = 4)
poly4 = PolynomialFeatures(degree=4)
X_poly_4 = poly4.fit_transform(X)

# 3️⃣ Fit the model
lin_reg = LinearRegression()
lin_reg.fit(X_poly_4, Y)

# 4️⃣ Predict values
Y_pred_poly4 = lin_reg.predict(X_poly_4)

# 5️⃣ Evaluate accuracy using R² score
accuracy = r2_score(Y, Y_pred_poly4)

print('Regression Score Function:', accuracy)
print(f'Regression Score Function: {int(accuracy * 100)}%')

# 6️⃣ Optional: plot results
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred_poly4, color='red', label='Polynomial Regression (Degree = 4)')
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('Hours vs Marks (Polynomial Regression Degree = 4)')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Position_Salaries.csv')
X = data[['Level']].values
y = data['Salary'].values

# Try polynomial degrees 1 to 5 (no X_grid)
plt.scatter(X, y, color='red', label='Actual Data')

for degree in range(1, 6):
    # Create polynomial features
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)

    # Fit linear regression
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly, y)

    # Predict using the same X (no grid)
    y_pred = lin_reg.predict(X_poly)

    # Plot the curve
    plt.plot(X, y_pred, label=f'Degree {degree}')

plt.title("Polynomial Regression (Without X_grid)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

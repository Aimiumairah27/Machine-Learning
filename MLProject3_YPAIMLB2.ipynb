```python
#OneHotEncoding

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(
    transformers=[("OneHot",OneHotEncoder(),[3])],
    remainder = 'passthrough'
)

x = transformer.fit_transform(x)
x = x.astype('int')
```

```python
# Load dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Startups_Data.csv')

# Define features (x) and target (y)
x = data[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]
y = data['Profit']

# Convert the 'State' column into numeric form (because it's text)
x = pd.get_dummies(x, drop_first=True)

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Split the dataset into 2 parts
# a.) Training Set    : apply a ML algo on Training Set
# b.) Testing Set      : we will be using in model evaluation

# 0.2 20% of the data will be used for Testing Purposes
# and remaining 80% of the data will be used for Training

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Original Shape
print(x.shape)
print(y.shape)
print('\nAfter Train Test Split\n')
print(x_train.shape)
print(x_train.shape)
print(y_train.shape)
print(y_train.shape)
```

```python
# Model Development
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
# save our trained model
import pickle
pickle.dump(mlr,open('multiple_reg.pkl','wb'))   # w --> write means create   b --> byte stream
```

```python
pred_values = mlr.predict(x_test)
pred_values
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, pred_values)
print('Score of Model :', score)

# r2 tells you how close your predicted numbers are to the real numbers overall
# r2 is a regression performance metrics
# r2 coefficient of determination
# r2 is a number between 0 and 1

# > 0.9        Excellent Fit
# 0.7 - 0.9    Good Fit
# 0.5 - 0.7    Moderate Fit
# <0.5         Weak Fit
```
